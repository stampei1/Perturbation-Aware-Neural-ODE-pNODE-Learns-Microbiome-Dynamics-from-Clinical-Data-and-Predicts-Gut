```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import models_and_functions as MyModels

    
df_antibiotics, antibiotic_types, df_relative_abundances= MyModels.wrangle_data()
df_relative_abundances = df_relative_abundances.drop_duplicates(subset = ['PatientID', 'DayRelativeToNearestHCT'])

enterococcus_infections = ['Enterococcus_Faecium',
                           'Enterococcus_Faecium_Vancomycin_Resistant',
                           'Enterococcus_Vancomycin_Resistant',
                           'Enterococcus_Faecalis']

df_infections = pd.read_csv('tblInfectionsCidPapers.csv')
df_bacilli_infections = df_infections[df_infections['InfectiousAgent'].isin(enterococcus_infections)]
df_bacilli_infections = df_bacilli_infections.drop('Timepoint', axis = 1).reset_index(drop = True)
df_bacilli_infections = df_bacilli_infections.sort_values(by=['PatientID', 'DayRelativeToNearestHCT'])
df_bacilli_infections = df_bacilli_infections.groupby("PatientID", as_index=False).first().drop('InfectiousAgent', axis = 1).reset_index(drop = True)
df_bacilli_abundances = df_relative_abundances.iloc[:,[2, -3,-2,-1]]
df_bacilli_infections
```

```python
from matplotlib.patches import Rectangle
from matplotlib.collections import LineCollection


def create_bacilli_timeline_plot(df_bacilli_abundances, df_bacilli_infections, random_seed=42):
    
    # Set random seed for reproducibility
    np.random.seed(random_seed)
    
    # Identify which patients had infections
    infected_patients = set(df_bacilli_infections['PatientID'].unique())
    all_patients = df_bacilli_abundances['PatientID'].unique()
    non_infected_patients = [p for p in all_patients if p not in infected_patients]
    
    # Separate and sort patients
    # For infected patients, sort by infection day
    # Only include patients with infections in the range -10 to 30
    infected_patient_list = []
    for patient_id in infected_patients:
        infection_day = df_bacilli_infections[
            df_bacilli_infections['PatientID'] == patient_id
        ]['DayRelativeToNearestHCT'].values[0]
        # Filter to only include infections in range -10 to 30
        if -10 <= infection_day <= 30:
            infected_patient_list.append((patient_id, infection_day))
    
    # Sort by infection day (earliest first)
    infected_patient_list.sort(key=lambda x: x[1])
    infected_patients_sorted = [p[0] for p in infected_patient_list]
    
    # Randomly select from each category
    if len(infected_patients_sorted) > 20:
        infected_patients_sorted = list(np.random.choice(infected_patients_sorted, 20, replace=False))
        # Re-sort by infection day after random selection
        infected_patient_list_subset = []
        for patient_id in infected_patients_sorted:
            infection_day = df_bacilli_infections[
                df_bacilli_infections['PatientID'] == patient_id
            ]['DayRelativeToNearestHCT'].values[0]
            infected_patient_list_subset.append((patient_id, infection_day))
        infected_patient_list_subset.sort(key=lambda x: x[1])
        infected_patients_sorted = [p[0] for p in infected_patient_list_subset]
    
    if len(non_infected_patients) > 20:
        non_infected_patients = list(np.random.choice(non_infected_patients, 20, replace=False))
    
    # Combine: infected patients first, then non-infected
    ordered_patients = infected_patients_sorted + non_infected_patients
    n_infected = len(infected_patients_sorted)
    n_total = len(ordered_patients)
    
    # Create figure with explicit size
    fig = plt.figure(figsize=(12, 9))
    ax = fig.add_subplot(111)
    
    # Color map for abundance (yellow to red)
    from matplotlib.colors import LinearSegmentedColormap
    colors_list = ['#FFFACD', '#FFE135', '#FFA500', '#FF6347', '#DC143C', '#8B0000']
    cmap = LinearSegmentedColormap.from_list('abundance', colors_list)
    
    # Plot each patient's timeline
    for idx, patient_id in enumerate(ordered_patients):
        y_position = n_total - idx - 1  # Reverse so first patient is at top
        
        # Get patient data, filter to time range -10 to 30
        patient_data = df_bacilli_abundances[
            df_bacilli_abundances['PatientID'] == patient_id
        ].sort_values('DayRelativeToNearestHCT')
        
        # If patient had infection, get infection day and filter data to before infection
        if patient_id in infected_patients:
            infection_day = df_bacilli_infections[
                df_bacilli_infections['PatientID'] == patient_id
            ]['DayRelativeToNearestHCT'].values[0]
            # Only show data before or at infection day
            patient_data = patient_data[patient_data['DayRelativeToNearestHCT'] <= infection_day]
        
        # Filter to time range
        patient_data = patient_data[
            (patient_data['DayRelativeToNearestHCT'] >= -10) & 
            (patient_data['DayRelativeToNearestHCT'] <= 30)
        ]
        
        days = patient_data['DayRelativeToNearestHCT'].values
        abundances = patient_data['Bacilli_relative'].values
        
        # Draw timeline (horizontal line)
        if len(days) > 0:
            ax.plot([max(-10, days.min()), min(30, days.max())], [y_position, y_position], 
                   'k-', linewidth=0.5, alpha=0.3, zorder=1)
        
        # Draw circles for each sample
        for day, abundance in zip(days, abundances):
            color = cmap(abundance)
            ax.scatter(day, y_position, s=120, c=[color], 
                      edgecolors='black', linewidths=0.5, zorder=3)
        
        # Add X marker if patient had infection
        if patient_id in infected_patients:
            infection_day = df_bacilli_infections[
                df_bacilli_infections['PatientID'] == patient_id
            ]['DayRelativeToNearestHCT'].values[0]
            
            ax.text(infection_day, y_position, 'X', 
                   fontsize=20, fontweight='bold', color='red',
                   ha='center', va='center', zorder=4)
    
    # Set x-axis limits
    ax.set_xlim(-10, 30)
    
    # Add separator line between infected and non-infected groups
    if n_infected > 0 and n_infected < n_total:
        separator_y = n_total - n_infected - 0.5
        ax.axhline(y=separator_y, color='gray', linestyle='--', 
                  linewidth=2, alpha=0.5, zorder=0)
        
        # Add group labels
        ax.text(ax.get_xlim()[0], n_total - n_infected/2, 
               'Patients with\nEnterococcus BSI', 
               fontsize=11, fontweight='bold', 
               ha='right', va='center',
               transform=ax.transData,
               bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.3))
        
        if len(non_infected_patients) > 0:
            ax.text(ax.get_xlim()[0], (n_total - n_infected - 1) / 2, 
                   'Non-infected\npatients', 
                   fontsize=11, fontweight='bold',
                   ha='right', va='center',
                   transform=ax.transData,
                   bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.3))
    
    # Formatting
    ax.set_xlabel('Day Relative to Nearest HCT', fontsize=12, fontweight='bold')
    ax.set_ylabel('Patient', fontsize=12, fontweight='bold')
    # ax.set_title('A. Abundance of bacteria of the class Bacilli tends to increase in fecal samples\nprior to an Enterococcus bloodstream infection', 
    #             fontsize=13, fontweight='bold', pad=20)
    
    # Remove y-axis ticks (patient IDs not needed for visual interpretation)
    ax.set_yticks([])
    
    # Add colorbar
    sm = plt.cm.ScalarMappable(cmap=cmap, 
                               norm=plt.Normalize(vmin=0, vmax=1))
    sm.set_array([])
    cbar = plt.colorbar(sm, ax=ax, pad=0.02, aspect=30)
    cbar.set_label('Bacilli Relative Abundance', fontsize=11, fontweight='bold')
    
    # Grid
    ax.grid(axis='x', alpha=0.3, linestyle=':', zorder=0)
    ax.set_axisbelow(True)
    
    plt.tight_layout()
    return fig, ax

# Usage:
fig, ax = create_bacilli_timeline_plot(df_bacilli_abundances, df_bacilli_infections, random_seed = 10)
plt.savefig('bacilli_timeline_figure.eps', dpi=300,format='eps')
plt.show()
```

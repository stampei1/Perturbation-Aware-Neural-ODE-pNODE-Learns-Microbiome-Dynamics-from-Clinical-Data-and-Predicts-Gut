```python
import numpy as np
import random
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import pandas as pd

from torchdiffeq import odeint
import models_and_functions as MyModels
df_antibiotics, antibiotic_types, df_relative_abundances = MyModels.wrangle_data()
df_absolute_abundances, df_relative_abundances_with_absolute_abundance_measurements_available = MyModels.wrangle_data_with_qpcr_measurements()
```

# Make the GLV train test split based on the NODE train test split:
GLV train and test sets will be the subsets of the NODE train and test sets which have absolute abundance measurements available (qPCR data available).

```python
node_test_df = pd.read_csv('node_model_test_df.csv', index_col = 0)
node_train_df = pd.read_csv('node_model_train_df.csv', index_col = 0)
test_df = df_absolute_abundances[df_absolute_abundances['PatientID'].isin(node_test_df['PatientID'].unique())]
train_df = df_absolute_abundances[df_absolute_abundances['PatientID'].isin(node_train_df['PatientID'].unique())]
glv_model = MyModels.GLVmodel(df_antibiotics, antibiotic_types)
glv_model.train(train_df)
torch.save(glv_model, 'glv_model.pth')
test_df.to_csv('glv_model_test_df.csv', index = False)
train_df.to_csv('glv_model_train_df.csv', index = False)
MyModels.compute_scatter_and_r2_absolute_abundance_model(glv_model, test_df, plot = True)
```
